```python
import torch
from torch.utils.data import DataLoader
from torchvision import transforms,datasets
from PIL import Image
import numpy as np
import os 
from matplotlib import pyplot as plt

imgs_path="./PetImages"#Just download the cat vs dog dataset online and put it in a right place
storage=[]             #format:[[path,label]]
for x in os.listdir(imgs_path):
    print(x)
    for y in os.listdir(imgs_path+'/'+x):
        img=Image.open(imgs_path+'/'+x+'/'+y)
        img=np.array(img)
        if x=='Cat':     #0=Cat,1=Dog
            storage.append([str(imgs_path+'/'+x+'/'+y),0])
        if x=='Dog':
            storage.append([str(imgs_path+'/'+x+'/'+y),1])

import random 
random.shuffle(storage)  #Random arrange the data
```

```python
index=13  #Using pyplot to check the label of each photo.
img=Image.open(storage[index][0])


plt.imshow(np.array(img))
plt.title("Label: "+str(storage[index][1]))
plt.show()
```
